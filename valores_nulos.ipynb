### Importar librerías

```python
import pandas as pd 
import numpy as np 
```

### Crear diccionario y DataFrame

```python
dict = {'Col1':[1,2,3,np.nan],
 'Col2':[4,np.nan,6,7],
 'Col3':['a','b','c',None]}
```

```python
df = pd.DataFrame(dict)
```

```python
df
```

### Ver valores nulos

```python
df.isnull()
```

```python
df.isnull()*1
```

### Llenar valores nulos

```python
df.fillna('Missing')
```

```python
df.fillna(df.mean())
```

### Interpolar

```python
df.interpolate() #Interpolacion de una serie asumiendo un valor que deberia seguir
```

### Borrar valores nulos

```python
df.dropna()
```
