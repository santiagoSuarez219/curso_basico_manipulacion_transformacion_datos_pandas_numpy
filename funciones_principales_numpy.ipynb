### Concatenate

```python
a = np.array([[1,2],[3,4]])
b = np.array([5,6])
```

```python
np.concatenate((a,b),axis=0)
```

```python
a.ndim
b.ndim
```

```python
b = np.expand_dims(b,axis=0)
```

```python
np.concatenate((a,b),axis=0)
```

```python
b
```

```python
np.concatenate((a,b),axis=1)
```

```python
np.concatenate((a,b.T),axis=1)
```
